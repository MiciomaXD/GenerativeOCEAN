```python
import pandas as pd
```

```python
ds=pd.read_csv("dataset/data-final.csv", sep="\t")
ds.info()
print(len(ds))
```

```python
ds.dropna(inplace=True)
ds.info()
print(len(ds))
```

```python
ds.to_csv("dataset/ds_nona.csv",header=True,index=False)
```

```python
oceanDF=pd.DataFrame(columns=["o","c","e","a","n"])
oceanDF.info()
```

```python
print()
oceanDF["o"]=ds.loc[:,["OPN"+str(i) for i in range(1,11)]].mean(axis=1).tolist()
oceanDF["c"]=ds.loc[:,["CSN"+str(i) for i in range(1,11)]].mean(axis=1).tolist()
oceanDF["e"]=ds.loc[:,["EXT"+str(i) for i in range(1,11)]].mean(axis=1).tolist()
oceanDF["a"]=ds.loc[:,["AGR"+str(i) for i in range(1,11)]].mean(axis=1).tolist()
oceanDF["n"]=ds.loc[:,["EST"+str(i) for i in range(1,11)]].mean(axis=1).tolist()
oceanDF.info()
```

```python
oceanDF.to_csv("dataset/ocean1M.csv",header=True,index=False)
```
